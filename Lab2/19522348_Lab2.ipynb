## Exercise 2

```python
salesData = pd.read_csv("/content/job-market.csv")
```

## Fix column DataTypes

```python
salesData = salesData.dropna()
```

```python
salesData
```

```python
salesData['Requirement'].astype("str")
```

```python
salesData['Id'] = salesData['Id'].astype(int)
```

## Check and Clean Data

```python
salesData['Date'] = salesData['Date'].str.split("T", 1, expand=True)[0]
```

```python
salesData['Title'] = salesData['Title'].str.replace("|", "/")
```

```python
salesData['FullDescription'] = salesData['FullDescription'].str.replace("<p>", "").str.replace("</p>", "").str.replace("<strong>", "").str.replace("<em>", "")
```

```python
salesData['FullDescription'] = salesData['FullDescription'].str.replace("</strong>", "").str.replace("<br>", "")
```

```python
salesData['FullDescription'].str.replace("&nbsp", "")
```

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext

for each in salesData['FullDescription']:
  each = cleanhtml(each)
```
